```python
# Task : List in Python

### Task 1: Create a List
#Create a list named `fruits` containing `"apple"`, `"banana"`, and `"cherry"`
My_List=['apple','banana','Cherry']
print(My_List)
```

```python
### Task 2: Access the Second Item
#Print the second item of the `fruits` list.
print(My_List[1])
```

```python
### Task 3: Change List Item
##Change the value of the first item in `fruits` to `"blackberry"`.
My_List[0]='blackberry'
print(My_List)
```

```python
### Task 4: Add an Item to the End
#Add `"orange"` to the end of the `fruits` list.
My_List.append('orange')
print(My_List)
```

```python
### Task 5: Insert an Item
#Insert `"lemon"` at position 1 in the `fruits` list.
My_List.insert(0,'Lemon')
print(My_List)
```

```python
### Task 6: Remove an Item
#Remove `"banana"` from the `fruits` list.
My_List.remove('banana')
print(My_List)
```

```python
### Task 7: Pop an Item
#Pop the last item from the `fruits` list and print it.
My_List.pop()
print(My_List)
```

```python
### Task 8: Copy a List
##Create a new list named `more_fruits` by copying the `fruits` list.
Fruits=My_List.copy()
print(Fruits)
print(My_List)
```

```python
### Task 9: Join Two Lists
##Create a third list named `all_fruits` by joining `fruits` and `more_fruits`.
List1=[1,2,3,4,5]
List2=[]
for x in List1:
    List2.append(x)
print(List2)
    
```

```python
### Task 10: Find the List Length
##Print the length of the `all_fruits` list.
L1=[1,2,3,4,'a','b','c']
le=len(L1)
print(le)
```

```python
### Task 11: List Slicing
##Print the first three items of the `all_fruits` list.
all_fruits=['Apple','Banana','Orange','Kiwi','Grapes']
print(all_fruits[0:3])
```

```python
### Task 12: Sort the List
##Sort the `all_fruits` list alphabetically.
all_fruits.sort()
print(all_fruits)
```

```python
### Task 13: Reverse the List
##Reverse the `all_fruits` list.
all_fruits.reverse()
print(all_fruits)
```

```python
### Task 14: Count an Item
##Count how many times `"apple"` appears in the `all_fruits` list and print the count.
all_fruits=['Apple','Banana','Orange','Kiwi','Grapes','Apple']
all_fruits.count('Banana')


```

```python
### Task 15: Clear a List
#Clear the `more_fruits` list.
more_fruits=['Apple','Banana','Orange','Kiwi','Grapes','Apple']
print(more_fruits)
more_fruits.clear()
print(more_fruits)

```

```python
### Task 16: List Multiplication
##Create a new list named `numbers` that contains 4 zeros.
numbers=[0]*4
print(numbers)
```

```python
### Task 17: Nested Lists
##Create a nested list named `matrix` that contains three lists, each with three numbers: 1, 2, 3; 4, 5, 6; and 7, 8, 9.
matrix=[[1,2,3],[4,5,6],[7,8,9]]
print(matrix)
```

```python
### Task 19: List Comprehension
#Use a list comprehension to create a new list named `squared_numbers` containing the square of each number in the `numbers` list.
squared_numbers=[x* 2 for x in range(1,10)]
print(squared_numbers)
    

```

```python
### Task 20: Filtering with List Comprehension
Create a list named `even_numbers` containing only the even numbers from the `numbers` list (assuming `numbers` list is filled with integers from 1 to 10).
evennumbers=[x for x in range(1,10) if x%2==0]
print(evennumbers)
```
