```python
#1. **Create an empty set.**
#- Expected outcome: `set()`
Myset=set()
print(Myset)
```

```python
#2. **Add an element '5' to an existing set `{1, 2, 3}`.**
#- Expected outcome: `{1, 2, 3, 5}`
Myset={1, 2, 3}
Myset.add(5)
print(Myset)
```

```python
#3. **Remove an element '2' from the set `{1, 2, 3}`.** - Expected outcome: `{1, 3}`
Myset1={1,2,3}
Myset1.remove(2)
print(Myset1)
```

```python
#5. **Find the length of the set `{1, 2, 3, 4, 5}`.**
#- Expected outcome: `5`
Myset={1,2,3,4,5}
print(len(Myset))
```

```python
#6. **Clear all elements from the set `{1, 2, 3, 4, 5}`.**
#- Expected outcome: `set()`
Myset={1, 2, 3, 4, 5}
Myset.clear()
print(Myset)
```

```python
#7. **Create a set with elements `{3, 4, 5, 6, 7}` and find its length.**
#- Expected outcome: `5`
MySet={3, 4, 5, 6, 7}
print(len(MySet))
```

```python
#8. **Find the union of two sets `{1, 2, 3}` and `{3, 4, 5}`.**
#- Expected outcome: `{1, 2, 3, 4, 5}`
set1={1, 2, 3}
set2={3, 4, 5}
print(set1|set2)
```

```python
#9. **Find the intersection of two sets `{1, 2, 3}` and `{2, 3, 4}`.**
#- Expected outcome: `{2, 3}`
set1={1, 2, 3}
set2={2, 3, 4}
print(set1&set2)
```

```python
#10. **Find the difference between two sets `{1, 2, 3}` and `{2, 3, 4}`.**
#- Expected outcome: `{1}`
set1={1,2,3}
set2={2,3,4}
print(set1-set2)
```

```python
#11. **Find the symmetric difference between two sets `{1, 2, 3}` and `{3, 4, 5}`.**
#- Expected outcome: `{1, 2, 4, 5}`

set1={1,2,3}
set2={3,4,5}
print(set1^set2)
```

```python
#12. **Check if the set `{1, 2}` is a subset of `{1, 2, 3, 4, 5}`.**
#- Expected outcome: `True`
set1={1, 2, 3, 4, 5}
set2={1, 2}
print(set2.issubset(set1))
```

```python
#13. **Check if the set `{1, 2, 3, 4, 5}` is a superset of `{3, 4}`.**
#- Expected outcome: `True`
set1={1, 2, 3, 4, 5}
set2={3, 4}
print(set1.issuperset(set2))
```

```python
#14. **Check if two sets `{1, 2, 3}` and `{4, 5, 6}` are disjoint.**
#- Expected outcome: `True`
set1={1, 2, 3}
set2={4, 5, 6}
print(set1.isdisjoint(set2))
```

```python
#15. **Create a set comprehension that squares each number in the range from 1 to 5.**
#- Expected outcome: `{1, 4, 9, 16, 25}`
set1={x**2 for x in range(1,6)}
print(set1)
```

```python
#18. **Check if the set `{1, 2, 3}` is a proper subset of `{1, 2, 3, 4}`.**
#- Expected outcome: `True`
set1={1, 2, 3}
set2={1, 2, 3, 4}
if set1.issubset(set2):
    print(True)
else:
    print(False)
```

```python
#19. **Create a set with elements `{1, 2, 3}` and then add elements `{4, 5}` to it.**
# - Expected outcome: `{1, 2, 3, 4, 5}`
set1={1, 2, 3}
set2={4,5}
set1.update(set2)
print(set1)
```
