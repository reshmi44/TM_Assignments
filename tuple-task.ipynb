```python
#1. **Create a Tuple:**
#Create a tuple named `my_tuple` with the elements `1`, `2`, and `3`.
my_tuple=(1,2,3)
print(my_tuple)
```

```python
#14. **Compare Tuples:**
#Compare two tuples `(1, 2)` and `(1, 2, 3)` to see if they are equal and print the result.
tup1=(1,2)
tup2=(1,2,3)
if tup1==tup2:
    print("Equal")
else:
    print("Not Equal")
```

```python
#15. **Tuple to String:**
#Convert the tuple `('h', 'e', 'l', 'l', 'o')` into a string and print it.
tup=('h', 'e', 'l', 'l', 'o')
stri=" "
for stri in tup:
    print(stri,end="")
```

```python
#16. **Minimum and Maximum:**
#Find the minimum and maximum elements in the tuple `(5, 1, 8, 3)` and print them.mini=
mini=min(5,1,8,3)
maxi=max(5, 1, 8, 3)
print('Minimum number is {} and maximum number is {} '.format(mini,maxi))
```
