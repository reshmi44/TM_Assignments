```python
#1. Write a lambda function to multiply two numbers and test it with the numbers 4 and 5.
multi=lambda x,y:x*y
print(multi(4, 5))
```

```python
#2.Create a lambda function that takes a string as input and returns its first character.
str=lambda s:s[0]
print(str("Hello"))
```

```python
#3. Use a lambda function to find the square of the number 9.
sqr=lambda sq:sq**2
print(sqr(6))
```

```python
#4. Write a lambda function that takes two arguments, a and b, and returns the result of a - b. 
#Test it with the values a = 10 and b = 5.

sub=lambda a,b:a-b
print("The diff is :",sub(10,5))
```

```python
#5. Create a lambda function that checks if a given number is odd. Test it with the number 7.
odd=lambda n:n%2 !=0
print(odd(7))
```

```python
#6. Use a lambda function to convert a temperature from Fahrenheit to Celsius. 
#Test it with the temperature 68Â°F.

ct_to_f=lambda c:9/5 *c /2 +32
print("Temp(F):",ct_to_f(68))
```

```python
#7. Write a lambda function that takes a list of numbers as input and returns the sum of all the numbers in the list.
#Test it with the list [1, 2, 3, 4, 5].
sumlist=lambda list1:sum(list1)
list1=[1, 2, 3, 4, 5]
print(sumlist(list1))
```

```python
#9. Use a lambda function to calculate the area of a rectangle given its length and width. 
#Test it with length = 4 and width = 6.
area=lambda length,width:length*width
print(area(4,6))
```
