```python
person = {
    'name': 'John',
    'age': 25,
    'city': 'New York'
}
print(person)
```

```python
#1. **Add a New Key-Value Pair:** Add a key 'email' with the value 'john@example.com' to the 
#above dictionary and print the updated dictionary.
person.update({'email':'john@example.com'})
print(person)
```

```python
#2. **Update an Existing Value:** Change the 'age' value in the dictionary to 26 
#and print the updated dictionary.
person['age']=26
print(person)
```

```python
#3. **Remove a Key-Value Pair:** Remove the 'city' key from the dictionary and print the updated dictionary.
person.pop('city')
print(person)
```

```python
#4. **Access a Value:** Print the value associated with the key 'name' in the dictionary.
print(person.get('name'))

```

```python
#5. **Check if Key Exists:** Check if the key 'age' exists in the dictionary 
#and print a message indicating its presence or absence.
if 'age' in person:
    print("Value Exist")
else:
    print("Not exists")
```

```python
#6.**Iterate Over Keys:** Use a for loop to iterate over the keys of the dictionary and print each key.
for i in person.keys():
    print(i)
```

```python
#7. **Iterate Over Values:** Use a for loop to iterate over the values of the dictionary and print each value.
for i in person.values():
    print(i)
```

```python
#8. **Iterate Over Key-Value Pairs:** Use a for loop to iterate over the key-value pairs of the dictionary
#and print each pair.
for i in person.items():
    print(i)
```

```python
#9. **Dictionary Length:** Print the number of key-value pairs in the dictionary.
dict1=len(person)
print(dict1)
```

```python
#10. **Copy a Dictionary:** Create a copy of the dictionary and print the copied dictionary.
person1=person.copy()
print(person1)
```

```python
#11. **Merge Two Dictionaries:** Create another dictionary with keys 'profession' and 'hobbies',
#and merge it with the existing dictionary. Print the merged dictionary.
person2={'profession':'IT Professional',
        'hobbies':'Reading'}
print(person1|person2)

```

```python
#12. **Get Value with get():** Use the `get()` method to retrieve the value of 'name' from the dictionary 
#and print it. If the key does not exist, print 'Key not found'.
x=person1.get('name')
print(x)
person1.pop('name')
if 'name' in person1:
    print("name found")
else:
    print("Key not found")
```

```python
#13. **Remove an Item with pop():** Use the `pop()` method to remove the 'age' key from 
#the dictionary and print the removed value and the updated dictionary.
d = {'name': 'Henry', 'age': 31}
print(d.pop('age'))
print(d)
```

```python
#14**Remove the Last Item with popitem():** Use the `popitem()` method to remove the last key-value pair from the
#
d = {'name': 'Ivy', 'age': 24}
print(d.popitem())
  
```

```python
#15. **Create a Dictionary from Keys:** Given a list of keys `['a', 'b', 'c']`, 
#create a dictionary where each key maps to the value 0. Print the dictionary.
keys = ['a', 'b', 'c']
d = dict.fromkeys(keys, 0)
print(d)
```

```python
#16. **Set Default Value:** Use the `setdefault()` method to set a default value for a key 'color' that
#does not exist in the dictionary. Print the updated dictionary.
d = {'car': 'Honda'}
d.setdefault('color', 'Red')
print(d)
```

```python
#17. **Update Dictionary with update():** Use the `update()` method to add a key-value pair 
#'gender': 'male' to the dictionary. Print the updated dictionary.
d = {'name': 'Karen'}
d.update({'gender': 'male'})
print(d)
```
