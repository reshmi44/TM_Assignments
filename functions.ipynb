```python
#5.**Function with Keyword Arguments**: Write a function called `describe_person` that takes two arguments,
#`name` and `age`, and prints a description of the person, like "John is 25 years old."
def describe_person(name,age):
    print("{} is {} years old".format(name,age))
name=input("Enter the name ")
age=input("Enter the Age")

describe_person(name,age)
```

```python
#6. **Function with Variable Number of Arguments**: Write a function called `sum_all` that takes any number of 
#arguments and returns their sum.
def sum_all(*args): 
    s=0 
    for i in args: 
        s=s+i 
    print(s) 

sum_all(20,30,5,10)
```

```python
#7. **Function as an Argument**: Write a function called `apply_operation` that takes three arguments: 
#a function `operation`, and two numbers `a` and `b`. 
#The function should apply the `operation` to `a` and `b` and return the result.
def apply_operation(a,b,c):
    if c=='+':
        return a+b
    elif c=='-':
        return a-b
    elif c=='*':
        return a*b
    elif c=='/':
        return a/b
    else:
        print("Wrong Operator")
    
a=int(input("Enter the number a"))
b=int(input("Enter the number b"))
c=input("Enter the Operator c")

apply_operation(a,b,c)

```

```python
#8. **Recursive Function**: Write a recursive function called `factorial` that takes a number `n` as a 
#parameter and returns the factorial of `n` (i.e., `n!`).
def factorial(n):
    if n==1:
        return 1
    else:
        return (n*factorial(n-1))

#n=int(input("enter the number"))            

factorial(4)
```

```python
#10. **Function Returning Function**: Write a function called `create_adder` that takes a number `x` as a
#parameter and returns a function that takes another number `y` as a parameter
#and returns the sum of `x` and `y`.
def create_adder(x):
    def adder(y):
        return x+y
    return adder
add6=create_adder(8)
print(add6(3))


```
