```python
#15. **BMI Calculator**:
#Write a program to calculate the Body Mass Index (BMI) and categorize it as underweight, normal, overweight, or obese.
height=float(input("Enter the Height"))
weight=float(input("Enter the Weight"))
BMI=weight/float(height*height)
if BMI<18:
    print("BMI is {} and is Underweight".format(BMI))
elif BMI>=18 and BMI<25:
    print("BMI is {} and is Normal ".format(BMI))
elif BMI>=25 and BMI<=30:
    print("BMI is {} and is Overweight",format(BMI))
else:
    print("Obese")
 
```

```python
#16. **Simple Calculator**
#Write a program to perform a simple calculation (addition, subtraction, multiplication, division) based on user input.
num1=int(input("Enter number1"))
num2=int(input("Enter number 2"))
oper=input("Enter the operation to be performed like +,-,*,/,%")
if oper=="+":
    sum=num1+num2
    print("Addition of {} and {} is {}".format(num1,num2,sum))
elif oper=='-':
    sub=num1-num2
    print("Substraction of {} and {} is {}".format(num1,num2,sub))
elif oper=='*':
    mul=num1*num2
    print("Multiplication of {} and {} is {}".format(num1,num2,mul))
elif oper=='/':
    div=num1/num2
    print("Multiplication of {} and {} is {}".format(num1,num2,div))
else:
    print("Wrong operator")
```
